# Importing

```python
import spacy
from spacy_layout import spaCyLayout
```

# Processing a Document 

```python
# Create blank spaCy Pipeline for English
nlp = spacy.load("en_core_web_sm")
```

```python
# Create an instance of spaCyLayout pipeline
layout = spaCyLayout(nlp)
```

```python
# Process a document and create a spaCy Doc object
doc = layout("data\RG-50.030.0045_trs_en.pdf")
```

# Attributes of the spaCy Layout Doc Object

```python
# The text-based contents of the document 
print(doc.text)
```

```python
#markdown representation of in the document
print(doc._.markdown)
```

```python
# Table in the document and their extracted data
print(doc._.tables)
```

```python
# Document layouting pages and sizes
print(doc._.layout)
```

```python
doc._.layout.pages
```

```python
# Examine Page 3 (Index 2) Data
doc._.pages[2]
```

```python
# Print the lable, text, and layout of each section in Page 3
for section in doc._.pages[2][1]:
    print("Label:", section.label_)
    print("Text:", section.text)
    print("Layout:", section._.layout)
    print()
   
```

# Visualizing the Result

```python
# Import required libraries
import pypdfium2 as pdfium
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Load and convert the PDF page to an image
pdf = pdfium.PdfDocument("data/RG-50.030.0045_trs_en.pdf")
page_image = pdf[2].render(scale=1)  # Get page 3 (index 2)
numpy_array = page_image.to_numpy()

# Get page 3 layout and sections
page = doc._.pages[2]
page_layout = doc._.layout.pages[2]

# Create figure and axis with page dimensions
fig, ax = plt.subplots(figsize=(12, 16))

# Display the PDF image
ax.imshow(numpy_array)

# Add rectangles for each section's bounding box
for section in page[1]:
    layout = section._.layout
    # Create rectangle patch
    rect = Rectangle(
        (layout.x, layout.y),
        layout.width,
        layout.height,
        fill=False,
        color='blue',
        linewidth=1,
        alpha=0.5
    )
    ax.add_patch(rect)
    
    # Add text label at top of box
    ax.text(layout.x, layout.y, section.label_,
            fontsize=8, color='red',
            verticalalignment='bottom')

# Set title and display
ax.set_title('Page 3 Layout with Bounding Boxes')
ax.axis('off')  # Hide axes
plt.show()
```
